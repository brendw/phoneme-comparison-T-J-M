````python
#function to convert trad -> jp simiplified variant

def convert_trad_to_jp(chars):
    output = ''
    for char in chars:
        output += shinjitai_conversion_dict.get(char, char)
    return output
    
````

````python
#function to convert trad -> jp simiplified variant

def convert_trad_to_jp(chars):
    
    output = ''

    for char in chars:

        if char in shinjitai_conversion_dict:
            x = shinjitai_conversion_dict[char]
            output += x
        else:
            output += char
       
    return output
    
````

# master dictionary format
Defintions
* hanzi: chinese character(s) corresponding to a word -- note that this project uses the traditional hanzi characters rather than the simplified characters
* kanji: chinese character(s) corresponding to a word as used in Japanese
* pinyin: romanization for pronunciation of mandarin
* tailo: romanization for pronunciation of taiwanese
* romanji: romanization for pronunciation of japanese
* hiragana: japanese alphabet for pronunciation of japanese

````python
#this function encodes the search term and searches jisho and returns the json entry

def search_jisho(word):
    
    url_encoded_keyword = urllib.parse.urlencode({'keyword': word}, encoding='utf-8')
    encoded_url = f'https://jisho.org/api/v1/search/words?{url_encoded_keyword}'
    
    #print(encoded_url)
    
    time.sleep(.5) #jisho has http too many requests error
    
    with urllib.request.urlopen(encoded_url) as url:
        json_entry = json.loads(url.read().decode()) 
    
    #print(json.dumps(json_entry, ensure_ascii=False, indent = 3)) 
    
    return json_entry
    
````

````python
#this function will retrieve hiragana readings from a json entry and return as a list
#returned list will be empty if no readings/entries

def get_j_reading( json_entry, j_search_query ):
    
    temp_reading_list = []
    
    for entry in json_entry["data"]: #searching through a list of dictionaries
        
        if j_search_query == entry["slug"]: #match is found
            
            readings = entry["japanese"] #readings is a list
            
            for item in readings: #item is a dict
                
                hiragana = item["reading"]
            
                if hiragana not in temp_reading_list:
                    temp_reading_list.append(hiragana) #add hiragana to list

            break #if match is found, stop searching the list
                
    return temp_reading_list
````

````python
#save jp readings in dictionary with key=trad char, value=list of readings
jp_kanji_readings_dict = {}
````

````python
#create first j_json_entry to pass to function

j_json_entry = search_jisho(convert_trad_to_jp(t_totalwords_list[0]))
````

````python
#this function takes the list of words from t_dict, converts from taiwanese hanzi to simplified jp
#and converts to url encoded format and searches in jisho to retrieve hiragana

def add_to_kanji_reading_dict( list_of_words, j_json_entry ):
    
    for t_entry in list_of_words: #current item from taiwanese dictionary

        j_search_query = convert_trad_to_jp(t_entry) #convert to jp simplified

        # moe dict organized by single char followed by compounds words starting with that char
        # search that word in jisho to get json_entry, and reuse that json_entry for the compound words

        if len(j_search_query) == 1: #this will create a new j_json_entry everytime a new single char comes up
            j_json_entry = search_jisho(j_search_query) 

        readings_list = get_j_reading( j_json_entry, j_search_query )

        if len(readings_list) > 0: #this will skip entries that don't have readings in jisho

            jp_kanji_readings_dict[t_entry] = readings_list 

            sharedwords_list.append(t_entry)
        
````

````python
#create a master dictionary with key,value format

master_dictionary = {}

for word in sharedwords_list:
    master_dictionary[word] = {"Mandarin": None, "Taiwanese": None, "Japanese": [None]}
    
#print(master_dictionary)
````

````python
#this function searches an entry in moe dict

#example moedict url: 
#https://www.moedict.tw/t/%E7%99%BC%E7%A9%8E.json 
#https://www.moedict.tw/t/發穎.json 

def search_moe_t( word ):
    
    time.sleep(0.5)
    
    encoded_url_pre = 'https://www.moedict.tw/t/{url_suffix}.json'\
    .format(url_suffix=urllib.parse.urlencode({"keyword": word}, encoding='utf-8'))
    
    #remove "keyword=" from url
    encoded_url = encoded_url_pre[:25] + encoded_url_pre[33:]
    
    #print(t_encoded_url)

    with urllib.request.urlopen(encoded_url) as url:
        json_entry = json.loads(url.read().decode())
        
    return json_entry
````

````python
#now using words common to both tw and jp dictionaries

#retrieve tai-lo

for word in sharedwords_list:
    
    try:

        t_json_entry = search_moe_t( word )

        for item in t_json_entry["h"]:
            t_reading = item["T"]

            #add tailo reading for each dictionary key
            master_dictionary[word]["Taiwanese"] = t_reading
    
    except:
        
        print(word, "tw error")
````

````python
#retrieve hiragana

#add hiragana for each dictionary key

# transfers the jp reading from the dict to the master_dictionary



for kanji in jp_kanji_readings_dict:

    master_dictionary[kanji]["Japanese"] = jp_kanji_readings_dict[kanji]

#i found a library that can convert hiragana->romaji
````

````python
#this function searches moe dict - mandarin and returns the json entry

#example moedict url: 
#https://www.moedict.tw/raw/%E8%90%8C
#https://www.moedict.tw/t/萌

def search_moe_m( word ):
    
    time.sleep(0.5)
    
    #encode hanzi with urllib.parse
    encoded_url_pre = 'https://www.moedict.tw/raw/{url_suffix}'\
    .format(url_suffix=urllib.parse.urlencode({"keyword": word}, encoding='utf-8'))
    
    #remove "keyword=" from url
    encoded_url = encoded_url_pre[:27] + encoded_url_pre[35:]

    with urllib.request.urlopen(encoded_url) as url:
        json_entry = json.loads(url.read().decode())
        
    return json_entry
````

````python
#this function separates multi character words to look up pronunciations separately

def multi_char_get_reading( word ): 
    temp_reading = ''

    count = 0
    for char in word:

        count +=1 

        if count > 1:
            temp_reading += ' ' #add space before subsequent characters if needed

        m_json_entry = search_moe_m( char )
        m_reading = m_json_entry["heteronyms"][0]["pinyin"]
        temp_reading += m_reading
    
    return temp_reading
````

````python
#retrieve pinyin

#reads the sharedwords_list and searches in mando dict and adds pronunciation to master_dictionary

no_compound_reading = []

for word in sharedwords_list:

    try:
        
        time.sleep(0.5)
        m_json_entry = search_moe_m( word )
        
        m_reading = m_json_entry["heteronyms"][0]["pinyin"]

        #add pinyin reading for each dictionary key
        master_dictionary[word]["Mandarin"] = m_reading
        
    except: 
        no_compound_reading.append(word)

        
for word in no_compound_reading:
    try: 
        m_reading = multi_char_get_reading(word)
        master_dictionary[word]["Mandarin"] = m_reading
    except:
        print(word, "doesnt exist in mandarin")
````

````python
#write master_dictionary to an output file

with open('master.txt', 'w') as outfile:
    json.dump(master_dictionary, outfile, ensure_ascii=False, indent=4)

#print(json.dumps( master_dictionary, ensure_ascii=False, indent=4))
````
